```python
import pandas as pd
knn_train = pd.read_csv(r'train_data/ExtraTreesRegressor_modified_with_bmi.csv')
knn_train
```

```python
df = pd.merge(knn_train, raw_data[['id', 'self_intro']], on='id', how='left')
df
```

```python
import re
import numpy as np
# 情緒字典標記函數
emotion_keywords = {
    1: ['boy', 'girlfriend', 'man','handsome','87','loser','dxxk','ggininder','fantastic','loser'],  # 男生關鍵詞列表
    -1: ['girl', 'pretty','beautiful','adorable','>.0','cuuuuuuute']  # 女生關鍵詞列表
}

def emotion_label(text):
    # 检查是否为有效字符串或字节对象
    if isinstance(text, str) or isinstance(text, bytes):
        

        # 分词
        words = re.findall(r'\b\w+\b', text)

        # 检查正面情绪关键词
        if any(word in emotion_keywords[1] for word in words):
            return 1

        # 检查负面情绪关键词
        if any(word in emotion_keywords[-1] for word in words):
            return -1
        # 中文检查
        if re.search(r'[\u4e00-\u9fff]+', text):
            return 1
    # 如果不是有效的字符串或字节对象，返回中性情绪标签
    return 0
```

```python
df['self_intro'] = df['self_intro'].str.lower()
df['emotion'] = np.vectorize(emotion_label)(df['self_intro'])
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
import numpy as np

# 假设df是您的DataFrame

# 准备特征和目标变量
X = df[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi', 'emotion']]
#X = df[['height', 'weight', 'sleepiness','bmi', 'emotion']]
y = df['gender'] - 1  # 确保类别标签从0开始

# 创建交叉验证器
cv = StratifiedKFold(n_splits=5)

# 特征预处理的管道，注意CatBoost不在这里
preprocessing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # 缺失值处理
    ('scaler', StandardScaler()),  # 特征缩放
    ('feature_selection', SelectKBest(f_classif, k='all')),  # 特征选择
])

# 用于存储每次交叉验证的准确率
scores = []

for train_index, test_index in cv.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # 拟合预处理管道
    X_train_transformed = preprocessing_pipeline.fit_transform(X_train, y_train)
    X_test_transformed = preprocessing_pipeline.transform(X_test)
    
    # 准备特征权重，假设emotion是最后一个特征
    feature_weights = [1] * (X_train_transformed.shape[1] - 1) + [1]  # 将emotion特征的权重设置为30倍
    
    # 定义CatBoost模型
    catboost = CatBoostClassifier(verbose=0, auto_class_weights='Balanced', feature_weights=feature_weights)
    
    # 拟合CatBoost模型
    catboost.fit(X_train_transformed, y_train)
    
    # 预测并计算准确率
    y_pred = catboost.predict(X_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)
    scores.append(accuracy)

# 打印每次交叉验证的准确率以及平均准确率
print('每次交叉验证的准确率:', scores)
print('平均准确率:', np.mean(scores))
```

```python
# 读取测试数据并预处理
knn_test = pd.read_csv(r'test_data/test_KNN_without_outlierBNI.csv')
knn_test['self_intro'] = knn_test['self_intro'].str.lower()
knn_test['emotion'] = np.vectorize(emotion_label)(knn_test['self_intro'])

# 提取需要的特征列
X_knn_test = knn_test[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi', 'emotion']]
#X_knn_test = knn_test[['height', 'weight', 'sleepiness','bmi', 'emotion']]

# 使用预处理管道对测试数据进行转换
X_knn_test_transformed = preprocessing_pipeline.transform(X_knn_test)

# 使用 CatBoost 模型进行预测
y_knn_pred = catboost.predict(X_knn_test_transformed)

# 添加预测结果到测试数据中
output = knn_test[['id']].copy()
output['gender'] = y_knn_pred + 1  # 因为模型中类别标签从0开始，需要加1得到实际类别

# 显示包含预测结果的测试数据
output.head(50)
```
