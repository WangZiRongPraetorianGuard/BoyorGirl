```python
df = pd.merge(knn_train, raw_data[['id', 'self_intro']], on='id', how='left')
df
```

```python
import re
import numpy as np
# æƒ…ç·’å­—å…¸æ¨™è¨˜å‡½æ•¸
emotion_keywords = {
    1: ['boy', 'girlfriend', 'man','handsome','87','loser','dxxk','ggininder','fantastic','loser'],  # ç”·ç”Ÿé—œéµè©åˆ—è¡¨
    -1: ['girl', 'pretty','beautiful','adorable','>.0','cuuuuuuute']  # å¥³ç”Ÿé—œéµè©åˆ—è¡¨
}

def emotion_label(text):
    # æ£€æŸ¥æ˜¯å¦ä¸ºæœ‰æ•ˆå­—ç¬¦ä¸²æˆ–å­—èŠ‚å¯¹è±¡
    if isinstance(text, str) or isinstance(text, bytes):
        

        # åˆ†è¯
        words = re.findall(r'\b\w+\b', text)

        # æ£€æŸ¥æ­£é¢æƒ…ç»ªå…³é”®è¯
        if any(word in emotion_keywords[1] for word in words):
            return 1

        # æ£€æŸ¥è´Ÿé¢æƒ…ç»ªå…³é”®è¯
        if any(word in emotion_keywords[-1] for word in words):
            return -1
        # ä¸­æ–‡æ£€æŸ¥
        if re.search(r'[\u4e00-\u9fff]+', text):
            return 1
    # å¦‚æœä¸æ˜¯æœ‰æ•ˆçš„å­—ç¬¦ä¸²æˆ–å­—èŠ‚å¯¹è±¡ï¼Œè¿”å›ä¸­æ€§æƒ…ç»ªæ ‡ç­¾
    return 0
```

```python
df['self_intro'] = df['self_intro'].str.lower()
df['emotion'] = np.vectorize(emotion_label)(df['self_intro'])
df
```

```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score


# å‡†å¤‡ç‰¹å¾å’Œç›®æ ‡å˜é‡
X = df[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi','emotion']]
y = df['gender'] - 1  # ç¡®ä¿ç±»åˆ«æ ‡ç­¾ä»0å¼€å§‹

# å®šä¹‰ç‰¹å¾åç§°åˆ—è¡¨ï¼Œä»¥ä¾¿åç»­æ‰“å°
feature_names = X.columns

# åˆ›å»ºäº¤å‰éªŒè¯å™¨
cv = StratifiedKFold(n_splits=5)

# åˆ›å»ºä¸€ä¸ªåŒ…å«é¢„å¤„ç†æ­¥éª¤å’ŒCatBoostæ¨¡å‹çš„ç®¡é“
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # ç¼ºå¤±å€¼å¤„ç†
    ('scaler', StandardScaler()),  # ç‰¹å¾ç¼©æ”¾
    ('feature_selection', SelectKBest(f_classif, k='all')),  # ç‰¹å¾é€‰æ‹©
    ('catboost', CatBoostClassifier(verbose=0, auto_class_weights='Balanced'))  # CatBoostæ¨¡å‹
])

# ç”¨äºå­˜å‚¨æ¯æ¬¡äº¤å‰éªŒè¯çš„å‡†ç¡®ç‡
scores = []

for train_index, test_index in cv.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # æ‹Ÿåˆç®¡é“
    pipeline.fit(X_train, y_train)
    
    # é¢„æµ‹å¹¶è®¡ç®—å‡†ç¡®ç‡
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    scores.append(accuracy)
    
    # è·å–é€‰ä¸­çš„ç‰¹å¾
    support = pipeline.named_steps['feature_selection'].get_support()
    selected_features = feature_names[support]
    
    print("Selected features:", selected_features)

# æ‰“å°æ¯æ¬¡äº¤å‰éªŒè¯çš„å‡†ç¡®ç‡ä»¥åŠå¹³å‡å‡†ç¡®ç‡
print('æ¯æ¬¡äº¤å‰éªŒè¯çš„å‡†ç¡®ç‡:', scores)
print('å¹³å‡å‡†ç¡®ç‡:', sum(scores) / len(scores))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# å‡è®¾dfæ˜¯æ‚¨çš„DataFrame

# å‡†å¤‡ç‰¹å¾å’Œç›®æ ‡å˜é‡
# X = df[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi', 'emotion']]
X = df[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'bmi', 'emotion']]
y = df['gender'] - 1  # ç¡®ä¿ç±»åˆ«æ ‡ç­¾ä»0å¼€å§‹

# åˆ›å»ºäº¤å‰éªŒè¯å™¨
cv = StratifiedKFold(n_splits=5)

# ç‰¹å¾é¢„å¤„ç†çš„ç®¡é“ï¼Œæ³¨æ„CatBoostä¸åœ¨è¿™é‡Œ
preprocessing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # ç¼ºå¤±å€¼å¤„ç†
    ('scaler', StandardScaler()),  # ç‰¹å¾ç¼©æ”¾
    ('feature_selection', SelectKBest(f_classif, k='all')),  # ç‰¹å¾é€‰æ‹©
])

# ç”¨äºå­˜å‚¨æ¯æ¬¡äº¤å‰éªŒè¯çš„å‡†ç¡®ç‡
scores = []

for train_index, test_index in cv.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # æ‹Ÿåˆé¢„å¤„ç†ç®¡é“
    X_train_transformed = preprocessing_pipeline.fit_transform(X_train, y_train)
    X_test_transformed = preprocessing_pipeline.transform(X_test)
    
    # å‡†å¤‡ç‰¹å¾æƒé‡ï¼Œå‡è®¾emotionæ˜¯æœ€åä¸€ä¸ªç‰¹å¾
    feature_weights = [1] * (X_train_transformed.shape[1] - 1) + [1]  # å°†emotionç‰¹å¾çš„æƒé‡è®¾ç½®ä¸º30å€
    
    # å®šä¹‰CatBoostæ¨¡å‹
    catboost = CatBoostClassifier(verbose=0, auto_class_weights='Balanced', feature_weights=feature_weights)
    
    # æ‹ŸåˆCatBoostæ¨¡å‹
    catboost.fit(X_train_transformed, y_train)
    
    # é¢„æµ‹å¹¶è®¡ç®—å‡†ç¡®ç‡
    y_pred = catboost.predict(X_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)
    scores.append(accuracy)

# æ‰“å°æ¯æ¬¡äº¤å‰éªŒè¯çš„å‡†ç¡®ç‡ä»¥åŠå¹³å‡å‡†ç¡®ç‡
print('æ¯æ¬¡äº¤å‰éªŒè¯çš„å‡†ç¡®ç‡:', scores)
print('å¹³å‡å‡†ç¡®ç‡:', np.mean(scores))
```
