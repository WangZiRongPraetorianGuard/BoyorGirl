```python
import pandas as pd
knn_train = pd.read_csv(r'train_data/ExtraTreesRegressor_modified_with_bmi.csv')
knn_train
```

```python
df = pd.merge(knn_train, raw_data[['id', 'self_intro']], on='id', how='left')
df
```

```python
import re
import numpy as np
# ÊÉÖÁ∑íÂ≠óÂÖ∏Ê®ôË®òÂáΩÊï∏
emotion_keywords = {
    1: ['boy', 'girlfriend', 'man','handsome','87','loser','dxxk','ggininder','fantastic','loser'],  # Áî∑ÁîüÈóúÈçµË©ûÂàóË°®
    -1: ['girl', 'pretty','beautiful','adorable','>.0','cuuuuuuute']  # Â•≥ÁîüÈóúÈçµË©ûÂàóË°®
}

def emotion_label(text):
    # Ê£ÄÊü•ÊòØÂê¶‰∏∫ÊúâÊïàÂ≠óÁ¨¶‰∏≤ÊàñÂ≠óËäÇÂØπË±°
    if isinstance(text, str) or isinstance(text, bytes):
        

        # ÂàÜËØç
        words = re.findall(r'\b\w+\b', text)

        # Ê£ÄÊü•Ê≠£Èù¢ÊÉÖÁª™ÂÖ≥ÈîÆËØç
        if any(word in emotion_keywords[1] for word in words):
            return 1

        # Ê£ÄÊü•Ë¥üÈù¢ÊÉÖÁª™ÂÖ≥ÈîÆËØç
        if any(word in emotion_keywords[-1] for word in words):
            return -1
        # ‰∏≠ÊñáÊ£ÄÊü•
        if re.search(r'[\u4e00-\u9fff]+', text):
            return 1
    # Â¶ÇÊûú‰∏çÊòØÊúâÊïàÁöÑÂ≠óÁ¨¶‰∏≤ÊàñÂ≠óËäÇÂØπË±°ÔºåËøîÂõû‰∏≠ÊÄßÊÉÖÁª™Ê†áÁ≠æ
    return 0
```

```python
df['self_intro'] = df['self_intro'].str.lower()
df['emotion'] = np.vectorize(emotion_label)(df['self_intro'])
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
import numpy as np

# ÂÅáËÆædfÊòØÊÇ®ÁöÑDataFrame

# ÂáÜÂ§áÁâπÂæÅÂíåÁõÆÊ†áÂèòÈáè
X = df[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi', 'emotion']]
#X = df[['height', 'weight', 'sleepiness','bmi', 'emotion']]
y = df['gender'] - 1  # Á°Æ‰øùÁ±ªÂà´Ê†áÁ≠æ‰ªé0ÂºÄÂßã

# ÂàõÂª∫‰∫§ÂèâÈ™åËØÅÂô®
cv = StratifiedKFold(n_splits=5)

# ÁâπÂæÅÈ¢ÑÂ§ÑÁêÜÁöÑÁÆ°ÈÅìÔºåÊ≥®ÊÑèCatBoost‰∏çÂú®ËøôÈáå
preprocessing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Áº∫Â§±ÂÄºÂ§ÑÁêÜ
    ('scaler', StandardScaler()),  # ÁâπÂæÅÁº©Êîæ
    ('feature_selection', SelectKBest(f_classif, k='all')),  # ÁâπÂæÅÈÄâÊã©
])

# Áî®‰∫éÂ≠òÂÇ®ÊØèÊ¨°‰∫§ÂèâÈ™åËØÅÁöÑÂáÜÁ°ÆÁéá
scores = []

for train_index, test_index in cv.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # ÊãüÂêàÈ¢ÑÂ§ÑÁêÜÁÆ°ÈÅì
    X_train_transformed = preprocessing_pipeline.fit_transform(X_train, y_train)
    X_test_transformed = preprocessing_pipeline.transform(X_test)
    
    # ÂáÜÂ§áÁâπÂæÅÊùÉÈáçÔºåÂÅáËÆæemotionÊòØÊúÄÂêé‰∏Ä‰∏™ÁâπÂæÅ
    feature_weights = [1] * (X_train_transformed.shape[1] - 1) + [1]  # Â∞ÜemotionÁâπÂæÅÁöÑÊùÉÈáçËÆæÁΩÆ‰∏∫30ÂÄç
    
    # ÂÆö‰πâCatBoostÊ®°Âûã
    catboost = CatBoostClassifier(verbose=0, auto_class_weights='Balanced', feature_weights=feature_weights)
    
    # ÊãüÂêàCatBoostÊ®°Âûã
    catboost.fit(X_train_transformed, y_train)
    
    # È¢ÑÊµãÂπ∂ËÆ°ÁÆóÂáÜÁ°ÆÁéá
    y_pred = catboost.predict(X_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)
    scores.append(accuracy)

# ÊâìÂç∞ÊØèÊ¨°‰∫§ÂèâÈ™åËØÅÁöÑÂáÜÁ°ÆÁéá‰ª•ÂèäÂπ≥ÂùáÂáÜÁ°ÆÁéá
print('ÊØèÊ¨°‰∫§ÂèâÈ™åËØÅÁöÑÂáÜÁ°ÆÁéá:', scores)
print('Âπ≥ÂùáÂáÜÁ°ÆÁéá:', np.mean(scores))
```

```python
# ËØªÂèñÊµãËØïÊï∞ÊçÆÂπ∂È¢ÑÂ§ÑÁêÜ
knn_test = pd.read_csv(r'test_data/test_KNN_without_outlierBNI.csv')
knn_test['self_intro'] = knn_test['self_intro'].str.lower()
knn_test['emotion'] = np.vectorize(emotion_label)(knn_test['self_intro'])

# ÊèêÂèñÈúÄË¶ÅÁöÑÁâπÂæÅÂàó
X_knn_test = knn_test[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi', 'emotion']]
#X_knn_test = knn_test[['height', 'weight', 'sleepiness','bmi', 'emotion']]

# ‰ΩøÁî®È¢ÑÂ§ÑÁêÜÁÆ°ÈÅìÂØπÊµãËØïÊï∞ÊçÆËøõË°åËΩ¨Êç¢
X_knn_test_transformed = preprocessing_pipeline.transform(X_knn_test)

# ‰ΩøÁî® CatBoost Ê®°ÂûãËøõË°åÈ¢ÑÊµã
y_knn_pred = catboost.predict(X_knn_test_transformed)

# Ê∑ªÂä†È¢ÑÊµãÁªìÊûúÂà∞ÊµãËØïÊï∞ÊçÆ‰∏≠
output = knn_test[['id']].copy()
output['gender'] = y_knn_pred + 1  # Âõ†‰∏∫Ê®°Âûã‰∏≠Á±ªÂà´Ê†áÁ≠æ‰ªé0ÂºÄÂßãÔºåÈúÄË¶ÅÂä†1ÂæóÂà∞ÂÆûÈôÖÁ±ªÂà´

# ÊòæÁ§∫ÂåÖÂê´È¢ÑÊµãÁªìÊûúÁöÑÊµãËØïÊï∞ÊçÆ
output.head(50)
```
