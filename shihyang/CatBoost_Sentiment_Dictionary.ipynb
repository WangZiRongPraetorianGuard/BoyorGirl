```python
import pandas as pd
knn_train = pd.read_csv(r'KNN_withBMI.csv')
knn_train
```

```python
df = pd.merge(knn_train, raw_data[['id', 'self_intro']], on='id', how='left')
df
```

```python
import re
import numpy as np
# ÊÉÖÁ∑íÂ≠óÂÖ∏Ê®ôË®òÂáΩÊï∏
emotion_keywords = {
    1: ['boy', 'girlfriend', 'man','handsome','87','loser','dxxk','ggininder','fantastic','loser'],  # Áî∑ÁîüÈóúÈçµË©ûÂàóË°®
    -1: ['girl', 'pretty','beautiful','adorable']  # Â•≥ÁîüÈóúÈçµË©ûÂàóË°®
}

def emotion_label(text):
    # Ê£ÄÊü•ÊòØÂê¶‰∏∫ÊúâÊïàÂ≠óÁ¨¶‰∏≤ÊàñÂ≠óËäÇÂØπË±°
    if isinstance(text, str) or isinstance(text, bytes):
        # ‰∏≠ÊñáÊ£ÄÊü•
        if re.search(r'[\u4e00-\u9fff]+', text):
            return 1

        # ÂàÜËØç
        words = re.findall(r'\b\w+\b', text)

        # Ê£ÄÊü•Ê≠£Èù¢ÊÉÖÁª™ÂÖ≥ÈîÆËØç
        if any(word in emotion_keywords[1] for word in words):
            return 1

        # Ê£ÄÊü•Ë¥üÈù¢ÊÉÖÁª™ÂÖ≥ÈîÆËØç
        if any(word in emotion_keywords[-1] for word in words):
            return -1

    # Â¶ÇÊûú‰∏çÊòØÊúâÊïàÁöÑÂ≠óÁ¨¶‰∏≤ÊàñÂ≠óËäÇÂØπË±°ÔºåËøîÂõû‰∏≠ÊÄßÊÉÖÁª™Ê†áÁ≠æ
    return 0
```

```python
df['self_intro'] = df['self_intro'].str.lower()
df['emotion'] = np.vectorize(emotion_label)(df['self_intro'])
df
```

```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score


# ÂáÜÂ§áÁâπÂæÅÂíåÁõÆÊ†áÂèòÈáè
X = df[['star_sign', 'phone_os', 'height', 'weight', 'sleepiness', 'iq', 'fb_friends', 'yt', 'bmi','emotion']]
y = df['gender'] - 1  # Á°Æ‰øùÁ±ªÂà´Ê†áÁ≠æ‰ªé0ÂºÄÂßã

# ÂÆö‰πâÁâπÂæÅÂêçÁß∞ÂàóË°®Ôºå‰ª•‰æøÂêéÁª≠ÊâìÂç∞
feature_names = X.columns

# ÂàõÂª∫‰∫§ÂèâÈ™åËØÅÂô®
cv = StratifiedKFold(n_splits=5)

# ÂàõÂª∫‰∏Ä‰∏™ÂåÖÂê´È¢ÑÂ§ÑÁêÜÊ≠•È™§ÂíåCatBoostÊ®°ÂûãÁöÑÁÆ°ÈÅì
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Áº∫Â§±ÂÄºÂ§ÑÁêÜ
    ('scaler', StandardScaler()),  # ÁâπÂæÅÁº©Êîæ
    ('feature_selection', SelectKBest(f_classif, k='all')),  # ÁâπÂæÅÈÄâÊã©
    ('catboost', CatBoostClassifier(verbose=0, auto_class_weights='Balanced'))  # CatBoostÊ®°Âûã
])

# Áî®‰∫éÂ≠òÂÇ®ÊØèÊ¨°‰∫§ÂèâÈ™åËØÅÁöÑÂáÜÁ°ÆÁéá
scores = []

for train_index, test_index in cv.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # ÊãüÂêàÁÆ°ÈÅì
    pipeline.fit(X_train, y_train)
    
    # È¢ÑÊµãÂπ∂ËÆ°ÁÆóÂáÜÁ°ÆÁéá
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    scores.append(accuracy)
    
    # Ëé∑ÂèñÈÄâ‰∏≠ÁöÑÁâπÂæÅ
    support = pipeline.named_steps['feature_selection'].get_support()
    selected_features = feature_names[support]
    
    print("Selected features:", selected_features)

# ÊâìÂç∞ÊØèÊ¨°‰∫§ÂèâÈ™åËØÅÁöÑÂáÜÁ°ÆÁéá‰ª•ÂèäÂπ≥ÂùáÂáÜÁ°ÆÁéá
print('ÊØèÊ¨°‰∫§ÂèâÈ™åËØÅÁöÑÂáÜÁ°ÆÁéá:', scores)
print('Âπ≥ÂùáÂáÜÁ°ÆÁéá:', sum(scores) / len(scores))
```
